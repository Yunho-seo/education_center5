```python
from matplotlib.patches import Polygon

def func(x):
  return (x-4) * (x-6) * (x-5) + 100  # x**3 : 3차 방정식 (변곡점이 2개)

a, b = 2, 9  # 적분 구간 
x = np.linspace(0, 10)
y = func(x)  # 높이값

fig, ax = plt.subplots()
ax.plot(x, y, "k", linewidth=2)
ax.set_ylim(bottom=0)

ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
# R에서 회귀선을 그릴 때, 폐다각형
poly = Polygon(verts, facecolor='green',  # 전경색 (green)
               edgecolor='0.5', alpha = 0.4)  # 경계선
ax.add_patch(poly)

# latex 문법 (수학기호를 출력)
ax.text(0.5 * (a+b), 30,  # 축의 좌표를 근거로 함 
        r'$\int_a ^ b f(x)\mathrm{d}x$',
        horizontalalignment = 'center',  # 수평 정렬
        fontsize=20)

fig.text(0.9, 0.05, '$x$')  # 전체를 0~1 사이즈로 보고, 위치를 지정한다.
fig.text(0.1, 0.9, '$x$')
ax.set_xticks((a, b))
plt.show()
```

```python
def f(x):
  return np.sin(x) + 0.5 * x  # sin 그래프 / 우상향

x = np.linspace(-2 * np.pi, 2 * np.pi, 50)  # -360 ~ 360 범위
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc = 0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
def fm(p):
  x, y = p
  return np.sin(x) + 0.25 * x + np.sqrt(y)  # x축, y축으로 상향하는 그래프 

x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)

X, Y = np.meshgrid(x, y)  # 20*20 그래프 -> 400개의 좌표
Z = fm((X, Y))
x = X.flatten()  # surface에 입력할 때에는 1차원으로 입력해야 함
y = Y.flatten()

from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(projection='3d')
# 면으로 출력
# row, columns 2칸씩 건너뛰기
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm, linewidth=0.5, antialiased=True)  # 픽셀은 사각형
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
def f(x, y):
  return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 - y**2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)  # 900개 좌표
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
ax.set_title('wireframe')
plt.show()
plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
plt.show()
```

```python
# 위경도 좌표로 변환
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent = 'South Korea')
def geocoding(address):
  try:
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]  # 위도, 경도 좌표 
    return x_y
  except:
    return [0, 0]
```

```python
import folium
centerpoint = geocoding('대전시청')
map_1 = folium.Map(location=centerpoint,
                   zoom_start=12,
                   tiles = 'Stamen Terrain')
folium.Marker([37.565711, 126.978090], popup='서울시청').add_to(map_1)

icon = ['house', 'cloud', 'cloud', 'cloud']
university = ['대전대학교', '충남대학교', '한밭대', '카이스트']
pos = []
for address in university:
  add = geocoding(address)
  pos.append(add)
print(pos)
i = 0
for address in pos:
  folium.Marker(address,
                popup = university[i],
                icon=folium.Icon(icon=icon[i])
              ).add_to(map_1)
  i += 1
map_1
```
